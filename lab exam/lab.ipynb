```python
import csv
import json
import re
```

```python
def Omega_load_and_integrate():
    fhzoo = open('C://Users//ub14-glab-008//Downloads//data//zoo.csv', mode = 'r', encoding="utf-8", newline = "")
    fhclass = open('C://Users//ub14-glab-008//Downloads//data//class.csv', mode = 'r', encoding="utf-8", newline = "")
    fhmeta = open('C://Users//ub14-glab-008//Downloads//data//auxiliary_metadata.json', 'r')
    readerzoo = csv.reader(fhzoo)
    readerclass = csv.reader(fhclass)
    A = json.load(fhmeta)

    Z=[]
    C=[]
    for i in readerzoo:
        Z.append(i)
    for i in readerclass:
        C.append(i)
    fhzoo.close()
    fhclass.close()
    fhmeta.close()
    return Z, C, A
Zoo, Class, Metadata = Omega_load_and_integrate()
print(Zoo)
```

```python
def Omega_Name_Normalization(x):
    with open('C://Users//ub14-glab-008//Downloads//data//zoo.csv', mode = 'w', encoding="utf-8", newline = "") as fh:
        for i in x:
            i[0] = re.sub(r'[^A-Za-z0-9]', '', i[0])
        writer = csv.writer(fh)
        writer.writerows(x)
Omega_Name_Normalization(Zoo)
```

```python
def Omega_fix_json(Metadata):
    f=-1
    for i in Metadata:
        flip = 0
        f=f+1
        for k,v in i:
            if flip == 0:
                Metadata[f]['animal_name'] = Metadata[f].pop([k])
            if flip == 1:
                Metadata[f]['habitat_type'] = Metadata[f].pop([k])
            if flip == 2:
                Metadata[f]['conservation_type'] = Metadata[f].pop([k])
Omega_fix_json(Metadata)
jstring = json.dumps(Metadata)
fhmet = open('C://Users//ub14-glab-008//Downloads//data//auxiliary_metadata.json', 'w')
fhmet.write(jstring)
fhmet.close()           
```
